**Day 1: Understanding Function**

Theoretical Question

```python
# A function in Python is a reusable block of code designed to perform a specific task. Instead of writing the same code multiple times, you define it once as a function and then "call" it whenever needed

# Basic concept of function in python
# 1=creat function using def keyword
# 2=call  uses function name followed by parenthesis {}
# 3=parameter=variable listed inside the parenthesis in the function definition
# 4=argument= the actual value you  pass to a function when calling it
```

for example:-

```python
def greet(name):  # function definition with parameter 'name'
    return f"Hello, {name}!"  # function body
```

```python
greet("AKASH")  # calling the function with argument "Alice"
```

Practical assessment:-

```python

def compute_basic_states(sample_data):
    stats = {
        "maximum" : {},
        "minimum" : {},
        "avg" : {},
    }
    stats["minimum"] = (min(sample_data))
    stats["maximum"] = (max(sample_data))
    stats["avg"] = (sum(sample_data)/len(sample_data))
    return stats
```

```python
sample_data = [10, 20, 30, 40, 50]
print(compute_basic_states(sample_data))                        
```

Day 5: Real-World Application and Assessmen+

Theoretical Question

```python
# Robust error handling becomes absolutely critical in scenarios where failure could lead to data loss, security breaches, or a poor user experience. Let’s explore one such scenario:
```

Practical Assignment:

```python
import maths_tools as mt 
```

```python
try:
    with open("dataset.txt", "r") as f:
        lines = f.readlines()[1:]   # skip header line

    ages = []
    heights = []

    for line in lines:
        age, height = line.strip().split(",")

        # clean data (skip missing or invalid)
        if age.strip() == "" or height.strip() in ("", "NaN"):
            continue

        ages.append(int(age))
        heights.append(int(height))

    # calculate averages using math_tools
    avg_age = mt.divi(sum(ages), len(ages))
    avg_height = mt.divi(sum(heights), len(heights))

    print("Ages:", ages)
    print("Heights:", heights)
    print("Average Age:", avg_age)
    print("Average Height:", avg_height)

except FileNotFoundError:
    print("Error: dataset.txt file is missing")
except ValueError:
    print("Error: Invalid data in dataset")
```
