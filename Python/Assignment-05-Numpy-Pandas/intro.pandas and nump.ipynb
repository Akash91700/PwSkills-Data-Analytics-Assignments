# Task 7: Insights & Trends

```python
sns.scatterplot(x = "Hours_Studied", y = "Score", data = df ,hue="Subject")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.title("Relationship Between Study Time and Scores")
plt.grid(True)
plt.show()
```

**General trend:** In most cases, as hours studied increase, scores tend to rise  suggesting a positive correlation between study time and performance.

For example, a Math student who studied around 10 hours scored ~85, while another who studied only 6 hours scored ~55

**Exceptions:** 
There are several points where more study time does not lead to better scores.

In Science, a student who studied 10 hours scored ~59, which is lower than another student who studied only 8 hours and scored ~72.

Similarly, in English, one student studied 12 hours and scored ~78, while another who studied 10 hours scored slightly less (~74), showing only a small improvement

```python

filtered = df[(df['Hours_Studied'] < 10) & (df['Score'] > 80)]
print(filtered)

```

```python
filtered = df[(df['Hours_Studied'] > 12) & (df['Score'] < 70)]
print(filtered)
```

These two cases show that:
- 1. More study time  guaranteed high score
- 2. Less study time  guaranteed low score

***In other words, study time alone is not a reliable predictor of performance. This highlights the importance of:***
- Study quality over quantity
- Individual learning styles
- Subject difficulty and interest
- Mental and physical well-being
