**Python**

A loop in Python is a programming construct that allows you to repeat a block of code multiple times until a specific condition is met.

**There are mainly two types of loops in Python:**

* for loop – used to iterate over a sequence (like list, tuple, string, etc.).

* while loop – runs as long as a given condition is True.

**Two reasons why loops are used in programming:**

1. To avoid repetition of code:
Loops help in executing the same set of instructions multiple times without writing the code repeatedly.

2. To process collections of data efficiently:
Loops make it easy to go through elements of a list, array, or any data structure one by one for processing.

```python
num = int(input("Enter a number:"))       # Take input from the user and convert it to an integer
if num > 0:                               # Check if the number is greater than zero
    print("The number is positive.")      # Print message if the number is positive
elif num < 0:                             # Check if the number is less than zero
    print("The number is negative.")      # Print message if the number is negative
else:                                     # If the number is neither positive nor negative
    print("The number is zero.")          # Print message if the number is zero
print(num)                                # Print the entered number
```

```python
number = int(input("enter a number:"))       # Take input from user and convert it to an integer
if number % 2 == 0:                           # Check if the number is divisible by 2 (even)
    print(number, "is an even number.")       # Print message if the number is even
else:                                         # If the number is not divisible by 2
    print(number, "is an odd number.")        # Print message if the number is odd
```

```python
total = 0                          # Initialize a variable to store the sum

for num in range(1, 21):           # Loop through numbers from 1 to 20
    total += num                   # Add each number to the total

print("The sum is:", total)        # Print the final sum
```
