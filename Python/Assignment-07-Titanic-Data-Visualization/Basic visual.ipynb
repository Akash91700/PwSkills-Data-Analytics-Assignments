# Q2. Bar Chart

```python
# Count the number of passengers in each class
class_counts = titanic['class'].value_counts()
class_counts
```

```python
# Create a bar chart for passenger class
sns.countplot(x='class', data=titanic, color='red', edgecolor='black')

# Add title and labels
plt.title('Number of Passengers in Each Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')

# Show the plot
plt.show()
```

#  Q3. Boxplot

```python
# Create boxplot of Age across different passenger classes
sns.boxplot(x="pclass", y="age", data=titanic)

# Add labels and title
plt.title("Boxplot of Age Across Passenger Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Show the plot
plt.show()
```

* Insights:

1. First Class:

Median age is higher  passengers were generally older.

The age range is wide, but most are adults, with fewer children.

2. Second Class:

Median age is moderate, slightly lower than first class.

Theres a mix of adults and younger passengers.

3. Third Class:

Median age is lowest  more young passengers and children.

The spread (interquartile range) is wider, showing greater age diversity.

# Q4. Heatmap

```python
# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Select numerical columns
numerical_cols = titanic.select_dtypes(include='number')

# Compute correlation matrix
corr_matrix = numerical_cols.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features in Titanic Dataset')
plt.show()
```

# **Q5. Pie Chart**

```python
# Count the number of survivors (0 = did not survive, 1 = survived)
survival_counts = titanic['survived'].value_counts()

# Create labels
labels = ['Did Not Survive', 'Survived']

# Create pie chart
plt.pie(
    survival_counts,
    labels=labels,
    autopct='%1.1f%%',     # show percentages with 1 decimal
    colors=['lightcoral', 'skyblue'],
    startangle=90,         # rotate to start from top
    explode=(0.05, 0)      # slightly separate the first slice
)

plt.title("Proportion of Passengers Who Survived vs Did Not Survive")
plt.show()
```

# Q6. Scatterplot

```python
# Create scatter plot
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic, palette='coolwarm', edgecolor='black')

# Add labels and title
plt.title('Scatterplot of Age vs Fare, Colored by Survival Status', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Fare', fontsize=12)

# Show the plot
plt.show()
```

* Observations:

1. Gender distribution:

Both males and females are spread across all ages,
but females tend to have paid higher fares on average, especially among younger and middle-aged groups.

2. Class influence:

Higher fares (right/top region of the chart) are mostly females, which often corresponds to first-class passengers.

Males dominate the lower-fare region (third class).

3. Age trend:

Theres no strong relationship between age and fare  fares vary more by class than by age.

# Q7. Histogram

```python
# Create a histogram of Age
sns.histplot(titanic['age'], bins=30, color='purple', edgecolor='black')

# Add labels and title
plt.title('Distribution of Passenger Age', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)

# Show the plot
plt.show()
```

* Observation of Age Distribution:

1. Most passengers are between 20 and 40 years old.

2. Fewer passengers are very young (010) or very old (>60).

3. Slightly right-skewed (more younger adults than elderly).

# Q8. Line Chart

```python
# Group by 'age' and calculate average fare
average_fare_by_age = titanic.groupby('age')['fare'].mean().reset_index()

# Display the first few rows
print(average_fare_by_age.head())
```

```python
# Group dataset by age and calculate average fare
avg_fare_by_age = titanic.groupby('age')['fare'].mean().dropna()

# Plot line chart
plt.plot(avg_fare_by_age.index, avg_fare_by_age.values, marker='o', color='red')
plt.title("Average Fare Paid by Age")
plt.xlabel("Age")
plt.ylabel("Average Fare")
plt.grid(True)
plt.show()
```

* Observation / Pattern:

1. Younger passengers (especially children) tend to pay lower fares.

2. Middle-aged adults (around 3040) often pay higher fares, likely first-class passengers.

3. Older passengers show more variability in fares.
