Question 6 Basic Operations:


```python
def find_second_largest(numbers):
    # Remove duplicates by converting to a set
    unique_numbers = set(numbers)
    
    # Check if there are at least two unique numbers
    if len(unique_numbers) < 2:
        return None  # No second largest number
    
    # Remove the largest number
    unique_numbers.remove(max(unique_numbers))
    
    # The next max is the second largest
    return max(unique_numbers)
# Example usage
numbers = [10, 20, 4, 45, 99, 99, 45]
second_largest = find_second_largest(numbers)
if second_largest is not None:
    print(f"The second largest number is {second_largest}.")
else:
    print("There is no second largest number in the list.")
```

# Question 9 Removing Duplicates:

```python
def remove_duplicates(lst):
    unique_list = []
    for item in lst:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list
# Example usage
original_list = [1, 2, 2, 3, 4, 4, 5, 1, 6]
result = remove_duplicates(original_list)
print("List after removing duplicates:", result)
```

Question 12 Basic Operations:

```python
def count_word_frequency(text):
    # Convert text to lowercase for case-insensitive counting
    text = text.lower()
    # Split the text into words
    words = text.split()
    
    frequency = {}
    for word in words:
        # Remove punctuation from the word
        word = word.strip('.,!?";:')
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
    return frequency
# Example usage
input_text = "This is a test. This test is only a test."
freq = count_word_frequency(input_text)
print(freq)
```

Question 13 Nested Dictionaries:

```python
# Create a nested dictionary with student names and their scores
students_scores = {
    "Alice": {"Math": 85, "Science": 92, "English": 88},
    "Bob": {"Math": 78, "Science": 81, "English": 90},
    "Charlie": {"Math": 95, "Science": 89, "English": 93}
}

# Initialize variables to track the top student and highest total score
top_student = None
highest_score = 0

# Iterate through the dictionary to calculate total scores
for student, scores in students_scores.items():
    total_score = sum(scores.values())
    if total_score > highest_score:
        highest_score = total_score
        top_student = student

# Print the student with the highest total score
print(f"The student with the highest total score is {top_student} with a score of {highest_score}.")
```

Question 14 Dictionary Updates: 

```python
# Dictionary of employee names and their salaries
employee_salaries = {
    "John": 50000,
    "Emma": 60000,
    "Michael": 55000,
    "Sophia": 62000
}
# Get employee name and new salary from user input
employee_name = input("Enter the employee name whose salary you want to update: ")
new_salary_input = input("Enter the new salary: ")
# Validate and update salary if employee exists
if employee_name in employee_salaries:
    try:
        new_salary = float(new_salary_input)
        employee_salaries[employee_name] = new_salary
        print(f"Salary of {employee_name} has been updated to {new_salary}.")
    except ValueError:
        print("Invalid salary input. Please enter a numeric value.")
else:
    print(f"Employee '{employee_name}' not found.")
```
